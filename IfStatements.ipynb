# Using if statements with lists

```python
requested_toppings = ['mushrooms', 'green peppers', 'extra cheese']

for requestedTopping in requested_toppings:
    if requestedTopping == "green peppers":
        print("Sorry, we are out of green peppers riht now.")
    else:
        print("Adding " + requestedTopping + ".")

print("Finished making your pizza!")
```

```python
# Cheking that a list is not empty
salsaToppings = []

if salsaToppings:
    for salsaTop in salsaToppings:
        print("Adding" + salsaTop + ".")
    print("Finishing making your salsa")
else:
    print("Are you sure you want ketchup")
```

```python
# Using multiple Lists

avaiableDegrees = ["Maths", "Physics", "Oncology", "Computer Science"]

requestDegrees = ["Physics", "Oncology", "Computer Science"]

for requestDegree in requestDegrees:
    if requestDegree in avaiableDegrees:
        print("You have been acepted in " + requestDegree + ".")
    else:
        print("Sorry, you haven't accepted in: " + requestDegree)

print("Try your best!")
```
