```python
# Keyword Arguments

def describePet(animalType, petName):
    print("I have a " + animalType)
    print("\nMy " + animalType + "'s name is " + petName)

describePet(petName = "Dory",animalType = "Dog" )
```

# Return values

```python
# Returning a simple value

def formattedName(first, last):
    fullName = first + " " + last
    return fullName.title()

musician = formattedName("miguel", "lopez")
print(musician)
```

```python
# Making an argument optional

def formattedName(first, last, middle = ""):

    if middle:
        fullName = first + " " + middle + " " + last
    else:
        fullName = first + " " + last

    return fullName.title()

musician = formattedName("Miguel", "lopez", "Angel")
print(musician)

artist = formattedName("Angel", "Miguelo")
print(artist)
```

```python
# Returning a dictionary

def bperson(first, last, age = ""):
    person = {"first: ": first, "last": last,}

    if age:
        person["age"] = age
    return person

musician = bperson("Miguel", "Lopez", age = 2)
print(musician)
```

```python
# Using a function with a while loop

def formattedName(first, last):
    fullName = first + " " + last
    return fullName

while True:
    print("Print 'q' when you finish")
    print("\nPlease enter your name: ")
    fname = input("First name: ")
    lname = input("Last name: ")
    quit = input("Have you finish")

    if quit == "q":
        break

    formName = formattedName(fname, lname)
    print("\nHi, my name is: " + formName)
```

```python
# Try it yourself

def cityCountry(city, country):
    text = city + ", " + country
    return text

leonMexico = cityCountry("Leon", "Mexico")
print(leonMexico)

santiagoChile = cityCountry("Santiago", "Chile")
print(santiagoChile)

bostonUSA = cityCountry("boston", "USA")
print(bostonUSA)
```

```python
def makeAlbum(artistName, albumTitle, tracks = ""):
    album = {"Artist Name": artistName, "Album Title": albumTitle,}
    if tracks:
        album["tracks"] = tracks
    return album

mAlbum = makeAlbum("Siddhartha", "El pez")
print(mAlbum)

nAlbum = makeAlbum("Bad bunny", "YOLDL", 17)
print(nAlbum)
```

```python
# modifying a list in a function

def models(unprinted_designs, completed_models):
    while unprinted_designs:
        currentDesing = unprinted_designs.pop()
        print("These designs are complete: " + currentDesing)
        completed_models.append(currentDesing)

def showModels(completed_models):
    print("These models are complete")
    for model in completed_models:
        print(model)

unprinted_designs = ['iphone case', 'robot pendant', 'dodecahedron']
completed_models = []

models(unprinted_designs, completed_models)
showModels(completed_models)
```

```python
# try it yourself

def printMagicians(magicianss):
    for magicianss in magicians:
        print(magicianss)

magicians = ["mike", "miguelo", "el bicho"]
printMagicians(magicians)
```

```python
def makeGreat(magicians, make_great):
    while magicians:
        greatMagician = magicians.pop()
        promp = "The great "
        make_great.append(promp + greatMagician)

magicians = ["mike", "miguelo", "el bicho"]
make_great = []

makeGreat(magicians[:], make_great)
print(magicians)
print(make_great)

```

# Passing an arbitrary number of arguments

```python
def make_pizza(*toppings):
    print("\nMaking a pizza with the following toppings")
    for topping in toppings:
        print(topping)

make_pizza("peperoni")
make_pizza("mushrooms", "green peppers", "extra cheese")
```

```python
# Using arbitrary keyword arguments

def build_profile(first, last, **user_info):
    profile = {}
    profile["first name"] = first
    profile["last name"] = last
    for key, value in user_info.items():
        profile[key] = value
    return profile

user_profile = build_profile("albert", "einstein", location = "princeton",
                             field = "physics")

print(user_profile)
```

```python
def makeCar(manufacturer, modelName, **carInfo):
    car = {}
    car["manufacturer"] = manufacturer
    car["model name"] = modelName
    for key, value in carInfo.items():
        car[key] = value
    return car

car1 = makeCar("toyota", "subaru", age = 1990, color = "blue",
              country = "germany")

print(car1)
```

# Storing your functions in modules

```python
import pizza
pizza.make_pizza(15, "jamon")
# module_name.function_name()
```

```python
# importing specific functions
from pizza import make_pizza
make_pizza(34, "green peppers")
```

```python
# using as to give a function an alias

from pizza import make_pizza as mp

mp(16, "extra cheese")
```

```python
# using as to give a module an alias

import pizza as p
p.make_pizza(58, "chorizo")
```

```python
# importing all functions in a module

from pizza import *

make_pizza(99, "elote")

# from module_name import *
```
