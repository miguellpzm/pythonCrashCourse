# Introducing **Lists**

```python
# Accessing Elements in a List
bicycles = ["trek", "redline", "specialized"]
print(bicycles[0])
print(bicycles[0].title())
print(bicycles[1])
print(bicycles[2])

#last element in a list
print(bicycles[-1])
print(bicycles[-2])
```

```python
# Using individual values from a list
message = "My first bicycle was a " + bicycles[0].title() + "."
print(message)
```

```python
# Names
names = ["Carlos", "Jotape", "Pepon"]
print(names[0])
print(names[1])
print(names[2])
```

```python
# Greetings
message1 = "Hi, "
message2 = ". How are you?"
print(message1 + names[0] + message2)
print(message1 + names[1] + message2)
print(message1 + names[2] + message2)
```

```python
# Your Own List
ftransportation = ["metro", "helicopter", "bus"]
print("I like to use the" + ftransportation[0] + " in boston")
print("I wish to use a " + ftransportation[1] + " to move to vallarta")
print("I like the " + ftransportation[-1] + " in suiza")
```

```python
# Modifying elements in a list
ftransportation[-1] = "airplane"
print(ftransportation)

ftransportation.append("yate")
print(ftransportation)
```

```python
# Empty list
food = []

food.append("tacos")
food.append("wings")
food.append("pizza")

print(food)
```

```python
# Insterting elements into a list

companys = ["nokia", "nike", "adidas"]

companys.insert(0, "under armour")
print(companys)
```

```python
# Removing elements from a list

cities = ["leon", "londres", "madrid"]

del cities[0]
print(cities)
```

```python
# pop() method
popped_cities = cities.pop()
print(popped_cities)

print("The last citie that i visited is: " + popped_cities)
```

```python
# Popping items from any position in a list
maths = ["statistics", "algebra", "trigonometry"]
first_learned = maths.pop(1)
print("The first tipic that i learned in math was: " + first_learned)
```

```python
# Removing an item by value
physics = ["particle", "astrophysics", "medical physics"]

physics.remove("particle")
print(physics)

difficult = "astrophysics"
physics.remove(difficult)
print(physics)
```

```python
# Organizing a List
# .sort() method

cars = ["bmw", "audi", "toyota", "subaru"]
cars.sort()
print(cars)

cars1 = ["bmw", "audi", "toyota", "subaru"]
cars.sort(reverse = True)
print(cars)
```

```python
# sorted() function

cars = ["bmw", "audi", "toyota", "subaru"]

print(cars)

print(sorted(cars))
```

```python
# Reverse Order

cars.reverse()
print(cars)

cars.reverse()
print(cars)
```

```python
# Length of a list

len(cars)
```

```python
# Try it yourself

locations = ["Roma", "Tokyo", "Sevilla", "Nueva Zelanda", "Australia"]

print(locations)

print(sorted(locations))

print(locations)

print(sorted(locations, reverse = True))

print(locations)

locations.reverse()
print(locations)

locations.reverse()
print(locations)

locations.sort()
print(locations)

locations.sort(reverse = True)
print(locations)

print("The total of cities that I want to visit is: " + str(len(locations)))
```
