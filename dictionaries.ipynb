# Loopoing through a dictionary

```python
# Loopingg through all key-value pairs

physicist = {
    'username': 'efermi',
    'first': 'enrico',
    'last': 'fermi',
}

for u, v in physicist.items():
    print("\nKey: " + u)
    print("Value: " + v)
```

```python
favoriteLanguages = {
    'jen': 'python',
    'sarah': 'c',
    'edward': 'ruby',
    'phil': 'python',
    }

friends = ["phil", "sarah"]
for name in friends:
    print(name.title())

    if name in favoriteLanguages:
        print("Hi " + name.title() + " I see your favorite lenguage is " +
              favoriteLanguages[name].title() + "!")

if "mike" not in favoriteLanguages.keys():
    print("\nMike, please take our poll!")
```

```python
# Looping through a dictionary´s keys in order

for names in sorted(favoriteLanguages.keys()):
    print(names.title() + ", thank you for taking the poll")


# We use set() to identifie the unique items in the list
for languaje in set(favoriteLanguages.values()):
    print(languaje)
```

# Nesting

```python
# A list of dictionaries

alien_0 = {'color': 'green', 'points': 5}
alien_1 = {'color': 'yellow', 'points': 10}
alien_2 = {'color': 'red', 'points': 15}

aliens = [alien_0, alien_1, alien_2]

for alien in aliens:
    print(alien)
```

```python
alienss = []

for alienNumber in range(30):
    new_alien = {'color': 'green', 'points': 5, 'speed': 'slow'}
    alienss.append(new_alien)

for aliene in alienss[0:3]:
    if aliene["color"] == "green":
        aliene["color"] = "yellow"
        aliene["speed"] = "medium"
        aliene["points"] = 10
    elif aliene["color"] == "yellow":
        aliene["color"] = "red"
        aliene["speed"] = "fast"
        aliene["points"] = 15

for alienn in alienss[:5]:
    print(alienn)


print(len(alienss))
```

```python
# A dictionary in a dictionary

users = {
    'aeinstein': {
        'first': 'albert',
        'last': 'einstein',
        'location': 'princeton',
        },
    'mcurie': {
        'first': 'marie',
        'last': 'curie',
        'location': 'paris',
        },
    }

for username, info in users.items():
    print("\n Username: " + username)
    fullName = info["first"] + " " + info["last"]
    location = info["location"]

    print("Full name: " + fullName)
    print("location: " + location + "\n")
```

```python
# Try it yourself

people = {
   "carlos": {
        "name": "Carlos Uranga",
        "age": 21,
        "living": "monterrey"},
   "diego": {
        "name": "Diego Gonzalez",
        "age": 21,
        "living": "puebla",
    }

}

for friend, inff in people.items():
    namee = inff["name"]
    agef = inff["age"]
    livingf = inff["living"]

    print("name: " + namee)
    print("age: " + str(agef))
    print("living: " + livingf + "\n")
```

```python
pets = {
    "dory" : {
        "type" : "dog",
        "Owner" : "Mike",
    },
    "Mary" : {
        "type": "iguana",
        "Owner": "Miguelon"
    },
    "Doña Gata" : {
        "type": "cat",
        "Owner": "ito"
    }
}

for petName, infe in pets.items():
    namePet = petName
    typePet = infe["type"]
    ownerPet = infe["Owner"]

    print("name: " + namePet)
    print("type: " + typePet)
    print("Owner: " + ownerPet + "\n")


```
