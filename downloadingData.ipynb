# The CSV file format

```python
import csv
from datetime import datetime

#Get high temperatures
filename = "sitka_weather_2018_full.csv"
with open (filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)
    print(header_row)

    for index, column_header in enumerate(header_row):
        print(index, column_header)

    dates, highs, lows = [], [], []
    for row in reader:
        current_date = datetime.strptime(row[2], "%m/%d/%Y")
        dates.append(current_date)

        high = int(row[8])
        highs.append(high)

        low = int(row[9])
        lows.append(low)

    print(highs)
```

```python
# Plotting Data ing a temperature chart

from matplotlib import pyplot as plt

# plot data

fig = plt.figure(dpi = 128, figsize = (12, 6))
plt.plot(dates, highs, c = "red")
plt.plot(dates, lows, c = "blue")
plt.fill_between(dates, highs, lows, facecolor = "blue", alpha = 0.5)

# format plot

plt.title("Daily high and low temperatures in 2018 and some days of 2019", fontsize = 19)
plt.xlabel("", fontsize = 16)
fig.autofmt_xdate()
plt.ylabel("temperature (F)", fontsize = 16)
plt.tick_params(axis = "both", which = "major", labelsize = 16)

plt.show()
```
