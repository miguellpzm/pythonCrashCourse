```python
# title() method

name = "miguel lopez"
print(name.title())
```

```python
# upper and lower methods

name = "Miguel Lopez"
print(name.upper())
print(name.lower())
```

```python
# Combining or concatenating strings

first_name = "Miguel"
last_name = "Lopez"
full_name = first_name + " " + last_name
print("Hello, " + full_name.upper() + "!")

name_lower = "Hello, " + full_name.lower() + "!"
print(name_lower)
```

```python
#Adding whitespace to string with tabs or newlines
print("Python")
print("\tPython")
# to add a tab on the text: \t
print("lenguages: \nPython\nC\nJavaScript")
# to add a new line in a string: \n
```

```python
# Stripping Whitespace
favorite_language = "python "
favorite_language
# .rstrip() method
favorite_language.rstrip()
# .rstrip() to right and .lstrip() to the left
```
